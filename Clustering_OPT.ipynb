```python
# Hierarchical clustering. Documents are represented as tf-idf

from collections import defaultdict
import nltk
import math
import string
from nltk.stem.porter import *
from sklearn import linear_model


def read(f):
    for l in open(f):
        yield eval(l)

comments = []
usernames = []
vocab = set()
punctuation = set(string.punctuation)
N = 0

for i in read('opt_dataset/small_opt.json'):
    comments.append(i['comment'])
    usernames.append(i['userName'])
    N += 1

    
def getset(words):
    myset = set()
    for w in words:
        myset.add(w)
    return myset
        

idf2 = defaultdict(float) # inverse document frequency, initialised to zeros by default

for i in comments:
    r = ''.join([c for c in i.lower() if not c in punctuation])
    words = getset(r.split())
    for w in words:
        idf2[w] += 1
        
idf = defaultdict(float)        
for w in idf2:
    if(idf2[w] > 1):
        idf[w] = math.log(N/idf2[w])
        vocab.add(w)

traverse = defaultdict(int)
i = 0
for w in idf:
    traverse[w] = i
    i += 1
        
```

```python
X = [[0.0 for i in range(len(idf))] for j in range(len(comments))]
for i in range(len(comments)):
    r = ''.join([c for c in comments[i].lower() if not c in punctuation])
    words = getset(r.split())
    for w in words:
        X[i][traverse[w]] += idf[w]
```

```python
import sklearn
import numpy as np
from sklearn.cluster import AgglomerativeClustering
from sklearn.decomposition import PCA
pca = PCA(n_components = 50)
pca.fit(X)
Z = pca.transform(X)
```

```python
from sklearn.cluster import KMeans
X = np.array(X)
ward = KMeans(n_clusters = 10)
ward.fit(X)
labels = ward.labels_
```
